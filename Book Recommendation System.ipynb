# Popularity based Recommender System


```python
books.head()
```

```python
ratings.head()
```

```python
#Joining books and users ratings into one table
```

```python
books_with_ratings=ratings.merge(books,on='ISBN')
books_with_ratings.head()
```

```python
books_with_ratings.shape
```

```python
popular_df=books_with_ratings.groupby('Book-Title').agg(num_rating=('Book-Rating','count'),avg_rating=('Book-Rating','mean'))
```

```python
popular_df=popular_df.reset_index()
```

```python
popular_df
```

```python
popular_df.sort_values('num_rating',ascending=False)
```

```python
#popularity recommendation is based on the popularity of the book (ratings >300)
#It is based on the ratings it got

popular_df=popular_df[popular_df['num_rating']>300].sort_values('num_rating',ascending=False)
```

```python
popular_df
```

```python
# Shrinking the dataframe to top 50 Books everyone needs to read
popular_df=popular_df.head(50).reset_index()
```

```python
popular_df
```

```python
#Merging my top 50 books with their author, and URL

popular_df=popular_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_rating','avg_rating']]
```

```python
popular_df=popular_df.sort_values('avg_rating',ascending=False)
```

```python
popular_df
```

# Collaborative Filtering
Similar Books prediction based on user feedback 

```python
books_with_ratings.head()
```

```python
# Grouping based on user id tells the no. of books read by each user
```

```python
x=books_with_ratings.groupby('User-ID').count()
```

```python
x
```

```python
x.shape
```

```python
# select only users who atleast gave feedback to 200 books

x=x[x['ISBN']>200]
x=x.sort_values('ISBN')
x
```

```python
x.index
```

```python
power_users=x.index
power_users
```

```python
#filtering ratings with best users
```

```python
filtered_ratings=books_with_ratings[books_with_ratings['User-ID'].isin(power_users)]
```

```python
filtered_ratings
```

```python
# Consider the best Users (who gave atleast 200 books ratings group them based on book titles
y=filtered_ratings.groupby('Book-Title').count()
```

```python
y.sort_values('User-ID',ascending=False)
```

```python
y=y[y['User-ID']>=50]
y
```

```python
famous_books=y.index
famous_books
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
# pivot table giving the rating for each book by each user
# Books rows with user_id as column

pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt=pt.fillna(0)
```

```python
pt
```

```python
similarity_scores=cosine_similarity(pt)
```

```python
similarity_scores
```

```python
df_temp=pd.DataFrame(similarity_scores)
df_temp
```

```python
df_temp_name=df_temp
df_temp_name.index=pt.index
df_temp_name.columns=pt.index
df_temp_name
```

```python
pt.index
```

```python
def recommend(book_name):
    index=np.where(pt.index==book_name)[0][0]
    similar_items=sorted(list(enumerate(similarity_scores[index])),key=lambda x: x[1],reverse=True)[1:6]
    data=[]
    for i in similar_items:
        item=[]
        df_temp=books[books['Book-Title']==pt.index[i[0]]]
        item.extend(list(df_temp.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(df_temp.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(df_temp.drop_duplicates('Book-Title')['Image-URL-M'].values))
        data.append(item)
    return data
```

```python
recommend('1984')
```
